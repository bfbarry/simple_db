```python
def to_unsigned_char(c):
    return ord(c) % 256


def hash(s, k=33):
    hash = 5381;
    for c in s:
        hash = hash * 33 + to_unsigned_char(c) # same as hash * 33 + c because bitshifting is faster
    return hash;


hash('asd') % 10
```
